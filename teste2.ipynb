```python
import tensorflow as tf
import matplotlib.pyplot as plt
```

```python
def adjust_contrast(image, factor):

    # Normalizando os valores dos pixels
    image = tf.cast(image, tf.float32) / 255.0
    
    # Ajustando o contraste
    adjusted_image = (image - tf.reduce_mean(image)) * factor + tf.reduce_mean(image)
    
    # Limitando os valores dos pixels
    adjusted_image = tf.clip_by_value(adjusted_image, 0, 1)
    
    # Convertendo de volta para uint8
    adjusted_image = tf.cast(adjusted_image * 255, tf.uint8)
    
    return adjusted_image
```

```python
#transformando a imagem num tensor
image_path = 'dark-landscape-1562281.jpg'
image = tf.io.read_file(image_path)
image = tf.image.decode_image(image, channels=3)

# Ajustando o contraste
adjusted_image = adjust_contrast(image, 2)

# Plotando as imagens original e ajustada
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.title('Original')
plt.imshow(image)
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title('Contraste ajustado')
plt.imshow(adjusted_image)
plt.axis('off')

plt.show()
```
